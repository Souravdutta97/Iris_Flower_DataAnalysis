# Line Plot:

Plot a line graph showing the sepal length of the three species of Iris flowers. Use different colors to represent each species.

```python
line=mat.groupby("Species")
line=line["SepalLengthCm"].plot()
plt.legend()
plt.xlabel("Species")
plt.ylabel("SepalLengthCm")
plt.title("the sepal length of the three species of Iris flowers")
```

# Bar Plot:

Create a bar plot showing the average sepal width for each species.

```python
avg_sepal=mat.groupby("Species")["SepalWidthCm"].mean()
avg_sepal.plot(kind="bar",color='skyblue')
plt.ylabel("SepalWidthCm")
plt.title("showing the average sepal width for each species.")
```

# Scatter Plot:

Create a scatter plot showing the relationship between the petal length and petal width. Use different colors to represent each species.

```python
sns.scatterplot(x="PetalLengthCm",y="PetalWidthCm",hue="Species",data=mat)
plt.title("the relationship between the petal length and petal width")
```

# Histogram:

Plot a histogram showing the distribution of the petal widths. Use different colors to represent each species and display a legend.


```python
sns.histplot(x="PetalWidthCm",hue="Species",data=mat,bins=4)
plt.title("The distribution of the petal widths")
```

# Pie Chart:

Assume that the dataset represents the entire population of Iris flowers. Create a pie chart showing the proportion of each species in the
population.

```python
explode=[0,0.1,0.1]
mat["Species"].value_counts().plot(kind="pie",autopct="%1.0f%%",explode=explode)
plt.title(" the proportion of each species in the population")
plt.show()
```

# Box Plot

# Create a box plot showing the distribution of sepal length for each species.

```python
sns.boxplot(x="SepalLengthCm",y="Species",data=mat)
plt.title("Showing the distribution of sepal length for each species.")
```

# Subplot

Combine the scatter plot and the histogram into one figure by using subplots.


```python
plt.figure(figsize=(10,6))
plt.subplot(121)
sns.scatterplot(x="PetalLengthCm",y="PetalWidthCm",hue="Species",data=mat)

plt.subplot(122)
sns.histplot(x="PetalWidthCm",hue="Species",data=mat,bins=4)
```
